```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

df = pd.read_csv('datasets_19728_29747_WorldCupMatches.csv')

#Insepct dataframe
print(df.head())
```

```python
#Creating a column that shows the total goals scored
#efore we create the new column, we will look at the date types of the cloumns to ensure that the value are either floats or ints

print(df.dtypes)
df['Total Goals'] = df['Home Team Goals'] + df['Away Team Goals']
```

```python
#creating a visualization of how many goals were scored each year of the World Cup that was held between 1930-2014


sns.set_style('whitegrid')
sns.set_context('poster',font_scale = .5)

#creating a figure and axes 
plt.figure(figsize = [12,7])
ax = plt.subplot(1,1,1)
ax = sns.barplot(data = df, x = 'Year', y = 'Total Goals')

#getting rid of the .0 in the Year column in our database for formatting purpose
year_format = [1930 + x*4 for x in range(0,40)]
year_format.remove(1942)
year_format.remove(1946)
ax.set_xticklabels(year_format)

ax.set_title('Total Goals per Year')

plt.show()
```

```python
#Creating a box plot to visualize the distribution of the goal data per year

df_goals = df[['Year','Total Goals']]
sns.set_context('notebook',font_scale = 1.25)

#creating figure and axes
plt.figure(figsize=[12,7])
ax1 = sns.boxplot(data = df_goals, x = 'Year', y = 'Total Goals', palette = 'Spectral')
ax1.set_xticklabels(year_format)
ax1.set_title('Goal Distribution per Year')

plt.show()
```
